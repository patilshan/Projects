```python
%%bigquery airlines
select *
from `msba-test-1.dbt_spatil.flights_final`
```

```python
airlines.head()
```

```python
southwest = airlines[airlines.airline_name == "Southwest Airlines Co."]
```

```python
southwest.head()
```

```python
others = airlines[airlines.airline_name != "Southwest Airlines Co."]
```

```python
others.head()
```

```python
delay_percent = ((southwest.arrival_delay >= 15).sum()/(southwest.arrival_delay).count())
percent_delay = delay_percent*100
print("Southwest Airlines delay percentage:", percent_delay)
```

```python
avg_delay = (southwest.arrival_delay).mean()
print("Southwest's average arrival delay =",avg_delay)
others_avg_delay = (others.arrival_delay).mean()
print("Other Airlines average arrival delay =",others_avg_delay)
```

```python
control = (southwest.arrival_delay_caused_by_carrier > 0).sum()
no_control = (southwest.arrival_delay > 0).sum()

percentage = (control/no_control)*100
print(" % of delays within Southwest's control =",percentage)
```

```python
time = (southwest.departure_delay-southwest.arrival_delay).mean()
print("Time Southwest making up during a flight =",time)
```
